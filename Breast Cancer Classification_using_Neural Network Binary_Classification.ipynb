## Train Test Split

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```


## Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Creating the Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw

model.add(Dense(units=30,activation='relu'))

model.add(Dense(units=15,activation='relu'))

# As it ios a binary classification So we kept Sigmoid function 
model.add(Dense(units=1,activation='sigmoid'))

# For a binary classification problem
model.compile(loss='binary_crossentropy', optimizer='adam')
```

## Training the Model 

### Example One: Choosing too many epochs and overfitting!

```python
# https://stats.stackexchange.com/questions/164876/tradeoff-batch-size-vs-number-of-iterations-to-train-a-neural-network
# https://datascience.stackexchange.com/questions/18414/are-there-any-rules-for-choosing-the-size-of-a-mini-batch

model.fit(x=X_train, 
          y=y_train, 
          # We are choosing epochs as 600 to get overfit to our data
          epochs=600,
          validation_data=(X_test, y_test), verbose=1
          )
```

```python
# model.history.history
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
 model_loss
```

```python
model_loss.plot()
# We can see our data is overfitting SO we will be adjusting it using early stopping method 
```

## Example Two: Early Stopping

We obviously trained too much! Let's use early stopping to track the val_loss and stop training once it begins increasing too much!

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
# We wan tto reduce the losses so we will be selecting mode as min for accuracy we select it max
# Patience means it the machine will wait and check for 25 more epochs after getting the best epoch count
# Read documentation of Early Stopping and see we hv to monitor the val_loss
# Verbose just goes through the data 
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop])
# callbacks just brings the EarlyStop function as a form of a List and its iteration will give us the result
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

## Example Three: Adding in DropOut Layers

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.5))
# This dropout wll randomly reduce 50% neuron in use epoch to improve our model

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.5))
# Again for this layer too 0.2 to 0.5 is a good dropout unit

model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
# WE can see the itertion stopped at 137 th iteration 
# Now just repeat the steps for model loss plot graph 
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
# It performed even better than early stopping grt
```

# Model Evaluation

```python
predictions = model.predict_classes(X_test)
# For Keras we use model.predict_classes instead of just model.predict
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# https://en.wikipedia.org/wiki/Precision_and_recall
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
